```python
from sklearn.cluster import KMeans
import numpy as np
file_data = file_data.replace(',', '', regex=True)
x=file_data[['Open', 'High', 'Low']]
```

```python
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(x)
```

```python
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

print("Labels:", labels)
print("Centroids:", centroids)
```

```python
for i in range(4):
    print(f"Cluster {i}:")
    print(x[labels == i])

```

```python
labels = kmeans.labels_
file_data['cluster'] = labels
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(x)

for i in range(4):
    plt.scatter(X_pca[file_data['cluster'] == i, 0],
                X_pca[file_data['cluster'] == i, 1],
                label=f'Cluster {i}')

plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("K-Means Clusters (PCA Reduced)")
plt.legend()
plt.grid(True)

plt.show()
```

```python
cols = ['Open', 'High', 'Low', 'LTP']

for col in cols:
    file_data[col] = file_data[col].astype(str).str.replace(',', '')
    file_data[col] = pd.to_numeric(file_data[col], errors='coerce')
```

```python
file_data['profit'] = file_data['LTP'] - file_data['Open']
file_data['profitable'] = file_data['profit'] > 0
cluster_behavior = file_data.groupby('cluster')['profitable'].mean()
print(cluster_behavior)
```

```python
threshold = 0.3

def is_good_cluster(cluster_id):
    return cluster_behavior[cluster_id] > threshold
```

```python
new_data = [[344.05, 346.50, 338.0]] 

# Step A: find cluster
cluster_id = kmeans.predict(new_data)[0]
print(cluster_id )

# Step B: check cluster behavior
if is_good_cluster(cluster_id):
    print("Worth Investing")
else:
    print("Not Worth Investing")
```

```python
from sklearn.tree import DecisionTreeClassifier

X = file_data[['Open', 'High', 'Low']]
y = file_data['cluster']

tree = DecisionTreeClassifier(max_depth=3)
tree.fit(X, y)
```

```python
from sklearn.tree import export_graphviz
import graphviz

dot_data = export_graphviz(
    tree,
    out_file=None,
    feature_names=['Open','High','Low'],
    class_names=True,
    filled=True
)

graph = graphviz.Source(dot_data)
graph
```
